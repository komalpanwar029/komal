```python
ax = sns.countplot(x =  'Gender',data = df)
```

```python
ax = sns.countplot(x =  'Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False).sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Gender', y = 'Amount' ,data = sales_gen)
```

from above graphs we see that most of the buyers are females and even the purchasing power of females are greater than men

```python
ax = sns.countplot(data = df , x = 'Age Group', hue =  'Gender') #hue is use for diff.b/w F and M
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age =  df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount' ,data = sales_age)
```

```python
sales_state = df.groupby(['State'] , as_index=False).sum().sort_values(by= 'Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state , x = 'State', y= 'Orders')



```

```python
sales_state = df.groupby(['State'] , as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state , x = 'State', y= 'Amount')
```

from above graphe we can see that  most of the  orders and total sales/ amounts are up  maharashtra and karnataka respectively

```python
ax = sns.countplot(data = df , x = 'Marital_Status',  hue =  'Gender')
for bars in ax.containers :
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status'] , as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state , x = 'Marital_Status', y = 'Amount' )
```

```python
sns.set(rc= {'figure.figsize': (20,5)})
ax = sns.countplot(data = df , x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

from above graphs we can see that most og the buyers are working in IT ,healthcare and aviation sector

```python
sns.set(rc= {'figure.figsize': (20,5)})
ax = sns.countplot(data = df , x = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'] , as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,20)})
sns.barplot(data = sales_state , x = 'Product_Category', y = 'Amount' )
```
